```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
```

```python
cars = pd.read_csv("cars.csv")
cars
```

```python
fig = plt.figure(figsize=(8, 6), dpi=80)
x = cars["Price mln"]
y = cars["Fuel consumption l"]
plt.plot(x, y, color = "blue", linewidth = 2)
plt.xlabel("Price mln")
plt.ylabel("Fuel consumption l")
```

```python
fig = plt.figure(figsize=(8, 6), dpi=80)
x = cars["Max speed kmps"]
y = cars["Fuel consumption l"]
plt.scatter(x, y, color = "red", marker = "D")
plt.xlabel("Max speed kmps")
plt.ylabel("Fuel consumption l")
```

```python
fig = plt.figure(figsize=(8, 6), dpi=80)
x = cars["Max speed kmps"]
y = cars["Сarrying capacity t"]
plt.scatter(x, y, color = "green")
plt.xlabel("Max speed kmps")
plt.ylabel("Сarrying capacity t")
```

```python
fig = plt.figure(figsize=(8, 6), dpi=80)
speed = cars["Max speed kmps"]
plt.hist(speed, 5, color = "green")
plt.xlabel("Max speed kmps")
plt.ylabel("Quantity")
```

```python
fig = plt.figure(figsize=(8, 6), dpi=80)
sb.boxplot(x="Max speed kmps", y="Сarrying capacity t", data=cars)
plt.xlabel("Max speed kmps")
plt.ylabel("Сarrying capacity t")
```
